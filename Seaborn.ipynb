## Import Data (You need to be connected to te Internet to be able import the data)


```python
#Seaborn internally has a number of datasets. We shall be using one of them.
#To see the list of the datasets, type
print(sns.get_dataset_names())
```

```python
# We shall be using "car_crasher"
crashes=sns.load_dataset('car_crashes')
crashes
```

## DISTRIBUTION PLOTS
### 1. Distribution Plot

```python
sns.displot(crashes['not_distracted'], kde=True, bins=25)
```

### 2. Joint Plot

```python
sns.jointplot(x='speeding',y='alcohol', data=crashes, kind='hex') #, you can have 'kde', hex
```

### 3.Pair Plot

```python
sns.pairplot(crashes)
```

### 4. Rugplot

```python
sns.rugplot(crashes['total'])
```

# STYLING

```python
sns.set_style('ticks') #other options are "white", "darkgrid","whitegrid" and "dark"
plt.figure(figsize=(8,4))
sns.set_context('poster',font_scale=1.4) #other options are 'talk' and 'paper'
sns.jointplot(x='speeding',y='alcohol', data=crashes, kind='reg')
```

# Categorical Plots

### 1. Bar Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
#print(sns.get_dataset_names())
tips_df=sns.load_dataset('tips')
tips_df
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.std) #the estimator can be used to calculate mean, median and other statistical functions
```

### 2. Count Plot


```python

sns.countplot(x='sex', data=tips_df, orient=)
```

### 3. BoxPlot

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

### 4.Violin Plot

```python
#This is a combination of the Box Plot and KDE
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

### 5. Strip Plot

```python
sns.stripplot(x='day', y='total_bill',data=tips_df, jitter=True, hue='sex',dodge=True)
```

### 6. Swarm Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='black')
```

# Palette

```python
plt.figure(figsize=(8,6))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex', palette='afmhot')
plt.legend(loc=0) #Upper right-1, Upper left-2, lower left-3,lower left 4
```

### Facet Grid

```python
tips_df=sns.load_dataset('tips')
tips_df
tips_fg=sns.FacetGrid(tips_df, col='time',row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8)
```

```python
tips_df=sns.load_dataset('tips')
tips_df
tips_fg=sns.FacetGrid(tips_df, col='time',row='smoker')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg=sns.FacetGrid(tips_df, col='time',hue='smoker',height=4,aspect=1.3)
tips_fg.map(plt.scatter, 'total_bill', 'tip',edgecolor='w')
```

```python
kws=dict(s=50, linewidth=.5,edgecolor='w')
tips_fg=sns.FacetGrid(tips_df, col='sex',hue='smoker',height=4,aspect=1.3, hue_order=['Yes','No'],hue_kws=dict(marker=['^','v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip',**kws)
```

```python
att_df=sns.load_dataset('attention')
att_df
```

```python
att_fg=sns.FacetGrid(att_df,col='subject',col_wrap=5,height=1.5)
att_fg.map(plt.plot,'solutions','score',marker='.')
```
